```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        if k == 1:
            return head
        n = 0
        temp = head
        while temp:
            n += 1
            temp = temp.next
        if k == 2:
            pre = head
            ptr = pre.next
            post = ptr.next
            l = 0
            while pre.next and n - l >= 2:
                l += 2
                conn = pre
                ptr.next = pre
                if conn == head:
                    head = ptr
                if post:
                    pre = post
                    ptr = post.next
                    if ptr:
                        post = ptr.next
                    conn.next = ptr
                else:
                    pre.next = None
            if l < n:
                conn.next = pre
            return head
        pre = head
        ptr = pre.next
        post = ptr.next
        l = 0
        self_conn = pre
        while post and n - l >= k:
            i = 0
            l += k
            if pre != head:
                s_conn = pre
            while i < k - 2:
                ptr.next = pre
                pre = ptr
                ptr = post
                post = post.next
                i += 1
            ptr.next = pre
            if self_conn:
                head = ptr
                conn = self_conn
                self_conn = None
            else:
                conn.next = ptr
                conn = s_conn
            if post:
                pre = post
                ptr = pre.next
                if ptr:
                    post = ptr.next
            else:
                conn.next = None
        if l < n:
            conn.next = pre
        return head
```
